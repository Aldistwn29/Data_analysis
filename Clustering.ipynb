## K-means

```python
# library yg digunakan
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

- 'Matplotlib' dan 'Seaborn' digunakan untuk Visualisasi
- 'numpy' digunakan untuk komputasi numerik
- 'sklearn' digunakan untuk oprasi machine learning
- 'make_blobs' digunakan untuk membuat dataset sederhana
- 'Kmeans' ini merupakan sala satu algorithma machine learning tepatknya algorithma clustering"pengelompokkan".

```python
X, y= make_blobs(n_samples = 500,n_features=2, cluster_std= 0.60, random_state= 0)
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.show()
```

- **make_blobs**
- 'n_sample' digunakan untuk mengatur jumlah sample data.
- 'centrs' digunakan untuk menentukan jumlah pusat cluster.
- 'n_feature' digunakan untuk menentukan jumlah feature pada sample.
- 'cluster_std' digunakan untuk mengontrol sebaran data dari pusat.
- 'random_state' di gunakan untuk mengatur seed generator acak sehingga hasilnya dapat direproduksi.

- **Sctter Plot**
- 'X' adalah array 2 dimensi yg di hasilkan oleh metode make_blobs
- 'X[:,0]' berarti semua baris pertama dari array X'
- 'X[:,1]' bearti semua baris kedua dari array X'
- s = 50, berarti ukuran titik scatter plot sama dengan 50.
- show digunakan untuk menampilkan visualisasi.

```python
kmeans = KMeans(n_clusters=5)
kmeans.fit(X)
```

- **Kmeans**
- 'n_cluster' jumlah cluster yg ingin dibentuk. defaultnya adalah 8
- 'init' adalah metode untuk inisialitati centroid, terdapat 2 metode yaitu kmeans ++, random acak.
- 'n_init' adalah Jumlah inisialisasi yang berbeda untuk dilakukan. defaultnya adalah 10.
- 'max_iter' Jumlah maksimum iterasi yang akan dilakukan untuk satu kali run dari algoritma K-Means. Deafaultnya 300
- 'tol' Relatif tolerance dengan kriteria inersia untuk menyatakan konvergensi. defaultnya 1e-4.
- 'verbose' Level keluaran (verbosity mode). Jika disetel ke 1 atau lebih tinggi, ini akan mencetak informasi tambahan saat algoritma berjalan. default adalah 0.
- 'random_satate' Penentu (seed) untuk generator angka acak untuk inisialisasi centroid. Ini bisa berupa integer, atau instance dari RandomState.
- 'copy_x' Jika True, data X akan disalin; jika False, data X akan diubah secara in-place selama fitting. defaultnya True.
- 'algotihma' Algoritma yang digunakan untuk melakukan pembaruan dalam setiap iterasi. Pilihan yang tersedia adalah 'auto', 'full', atau 'elkan'.

- 'fit' digunakan untuk melatih dataset yg digunakan.

```python
y_kmeans = kmeans.predict(X)
plt.scatter(X[:,0], X[:,1], s=50, c=y_kmeans)
plt.show()
```

- '.predict' digunakan untuk memprediksi data setelah data di latih.
- **Sctter Plot**
- 'X' adalah array 2 dimensi yg di hasilkan oleh metode make_blobs
- 'X[:,0]' berarti semua baris pertama dari array X'
- 'X[:,1]' bearti semua baris kedua dari array X'
- s = 50, berarti ukuran titik scatter plot sama dengan 50.
- c=y_kmeans, berarti setiap lebel di kelompokan dengan nilai yg sama dan di warna dengan titik yg sama pula.
- show digunakan untuk menampilkan visualisasi.
